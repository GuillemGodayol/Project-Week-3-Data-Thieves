```python
#Import required libraries
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup #don't forget to pip install BS

def cleantags(x):
    s1 = re.sub("<h2>", "", x)
    s2 = re.sub("</h2>", "", s1)
    s3 = re.sub("<strong>", "", s2)
    s4 = re.sub("</strong>", "", s3)
    return s4
def cleantagsdigits(x):
    s1 = re.sub("<h2>\d*\.\s", "", x)
    s2 = re.sub("</h2>", "", s1)
    s3 = re.sub("<strong>", "", s2)
    s4 = re.sub("</strong>", "", s3)
    return s4
```

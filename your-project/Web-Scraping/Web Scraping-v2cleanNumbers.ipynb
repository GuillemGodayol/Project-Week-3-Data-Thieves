```python
def cleantags(x):
    s1 = re.sub("<h2>", " ", x)
    s2 = re.sub("</h2>", " ", s1)
    s3 = re.sub("<strong>", " ", s2)
    s4 = re.sub("</strong>", " ", s3)
    return s4
```

```python
def cleantagsdigits(x):
    s1 = re.sub("<h2>\d*\.\s", " ", x)
    s2 = re.sub("</h2>", " ", s1)
    s3 = re.sub("<strong>", " ", s2)
    s4 = re.sub("</strong>", " ", s3)
    return s4
```

```python
df["interim_name"] = df["string"].apply(cleantags)
df['rank'] = df['interim_name'].apply(lambda s:s.split('.')[0])
```

```python
df["interim_nameb"] = df["string2"].apply(cleantagsdigits)
df['rankb'] = df['interim_nameb'].apply(lambda s:s.split('.')[0])
```

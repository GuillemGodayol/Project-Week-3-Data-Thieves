```python
#Import required libraries
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup #don't forget to pip install BS
```

```python
#
html_doc = 'https://marketing4ecommerce.net/youtubers-mas-seguidos-de-espana-2019-50/'
page = requests.get("https://marketing4ecommerce.net/youtubers-mas-seguidos-de-espana-2019-50/")
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup.find_all('h2')
```

```python
df = pd.DataFrame(soup.find_all('h2'))
```

```python
df = df.iloc[5:] #removed the first 5 rows as garbage
```

```python
df.columns = ['raw_name']
```

```python
df = df.drop([39,53], axis=0)
```

```python
df['raw2'] = df['raw_name'] 
df
```

```python
df['string'] = df['raw_name'].astype('str')  #convert dtype to string
df['string2'] = df['raw2'].astype('str')  #convert dtype to string
#df['string'] = df['raw_name'].apply(lambda x: str(x))
```

```python
df_bup = df #create backup
```

```python
df["interim_name"] = df["string"].apply(lambda x:re.sub("<h2>", " ", x))
df["interim2_name"] = df["interim_name"].apply(lambda x:re.sub("</h2>", " ", x))
df["interim3_name"] = df["interim2_name"].apply(lambda x:re.sub("<strong>", " ", x))
df["interim4_name"] = df["interim3_name"].apply(lambda x:re.sub("</strong>", " ", x))
df['rank'] = df['interim4_name'].apply(lambda s:s.split('.')[0])
```

```python
df["interim_nameb"] = df["string2"].apply(lambda x:re.sub("<h2>\d*\.\s", "", x))
df["interim2_nameb"] = df["interim_nameb"].apply(lambda x:re.sub("</h2>", "", x))
df["interim3_nameb"] = df["interim2_nameb"].apply(lambda x:re.sub("<strong>", "", x))
df["interim4_nameb"] = df["interim3_nameb"].apply(lambda x:re.sub("</strong>", " ", x))
df['rankb'] = df['interim4_nameb'].apply(lambda s:s.split('.')[0])
```

```python
df['URL'] = 'https://marketing4ecommerce.net/youtubers-mas-seguidos-de-espana-2019-50/'
```

```python
#df_clean = df[['rank','interim4_name','URL', 'rankb']]
df_clean = df[['rank','interim4_name','rankb']]
```

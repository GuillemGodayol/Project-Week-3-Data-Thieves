```python
def transformtime(raw):
    if 'H' not in raw and 'M' not in raw and 'S' in raw:    #seconds only
        x1 = raw.strip('PTS')
        return int(x1)
    elif 'H' not in raw and 'M' in raw and 'S' not in raw:    #minutes only
        x1 = raw.strip('PTM')
        return int(x1) * 60
    elif 'H' not in raw and 'M' in raw and 'S' in raw:    #mins and seconds only
        x1 = raw.strip('PTS')
        x3 = x1.split("M")
        lst_num = [int(num) for num in x3]
        return lst_num[0] * 60 + lst_num[1]
    elif 'H' in raw and 'M' in raw and 'S' in raw:    #all 3  PT1H23M37S
        x1 = raw.strip('PTS')                                  
        x3 = x1.split("H")
        x4 = x3[1].split("M")
        return (int( x3[0] )* 60 * 60) +  (int( x4[0] ) * 60 ) + int( x4[1] )
    else:
        print('error')
```
